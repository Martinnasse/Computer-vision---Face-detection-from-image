### 1. Import packages



```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

#   
#   
### 2. Read and display the original image



```python
persons_of_interest = cv2.imread('persons_of_interest.jfif',0)
plt.imshow(persons_of_interest,cmap='gray')
```

#   
#   
### 3. Face detection



#### - We will use a pre-trained file proivided by openCV

```python
# Pre-trained file from opencv
pre_trained_file = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

def face_detection(image,pre_trained_file):
    
    # Get the cordinates of the rectangles where we have a face
    rectangle = pre_trained_file.detectMultiScale(image) 
    
    face=image.copy()
    # Include the rectangles in the image
    for (x,y,w,h) in rectangle: 
        cv2.rectangle(face, (x,y), (x+w,y+h), (255,255,255), 2) 
        
    return face
  
face_detected_image = face_detection(persons_of_interest,pre_trained_file)
plt.imshow(face_detected_image,cmap='gray')
```
